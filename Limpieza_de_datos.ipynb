```python
import pandas as pd
import plotly.express as px
```

```python
# Esta linea de codigo ordena al lenguaje que lea el archivo que le estamos pasando
#df => significa dataframe
df = pd.read_excel("Movilidad_U_Quindio.xlsx")
```

```python
#Verificamos que el archivo este correcto
df.shape
```

```python
#Nos muestra los datos de los ultimos valores
df.tail(20)
```

```python
#Nos indica los valores que mas se repiten
df["Nombre de la actividad General"].describe()
```

```python
#Mostramos los datos unicos
df["Nombre de la actividad General"].unique()
```

```python
#La linea de codigo funciona para limpiar la base de datos, eliminando cualquier columna que tenga un campo de dato erroneo. Sin embargo este no lo tiene asi que decidi eliminar
data = df.drop(columns=["Origen"])
```

```python
#vemos que eliminamos el campo Origen
data.info()
```

```python
data.shape
```

```python
data.columns
```

```python
#empezamos a normalizar los datos. Cambiamos el nombre de los datos porque tiene caracteres especiales
data.rename(columns={"Facultad/Dependencia": "Facultad_Dependencia",
                     "Programa/Oficina":"Programa_Oficina",
                     "Nombre de la actividad General":"Nombre_Actividad_General",
                     "Ciudad de Origen":"Ciudad_Origen",
                     "Institución de Origen":"Institucion_Origen"
                     }, inplace=True)
```

```python
data.columns
```

```python
data.to_excel("Movilidad_normalizado.xlsx", index=False)
print("archivo guardado")
```
